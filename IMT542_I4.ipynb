```python
# I4
# I picked the data sources used in this code
# Code was written and completed with assistance from Google Gemini

import pandas as pd

# Read a csv file from a url and print the first 5 rows
# Pros: easy to use with pandas, useful for structured data, uses common format (csv) which is compatible with many tools
# Cons: need to have an exact URL otherwise it will not work, no real-time server-side updates
def read_csv_from_url(url: str):
    try:
        # Read the CSV directly into a pandas DataFrame
        df = pd.read_csv(url)
        print(f"CSV from {url} successfully uploaded.")
        print("First 5 rows of the data:")
        display(df.head())
    except Exception as e:
        print(f"Error reading CSV from URL: {e}")

# Using a public CSV found on GitHub
csv_url = "https://gist.githubusercontent.com/kevin336/acbb2271e66c10a5b73aacf82ca82784/raw/e38afe62e088394d61ed30884dd50a6826eee0a8/employees.csv"
read_csv_from_url(csv_url)
```

```python
import requests
import json

# Get JSON data from an API and print a sample
# Pros: APIs can provide real-time information because they are updated on the backend
# Cons: APIs may restrict the number of requests you can make, often requires keys for access, and has more complex error scenarios
def get_json_from_api(api_endpoint: str):
    try:
        response = requests.get(api_endpoint)
        response.raise_for_status()  # Raise an exception for HTTP errors (4xx or 5xx)
        data = response.json()

        print(f"Successfully fetched JSON from {api_endpoint}.")
        print("Sample of the data (first few items/keys):")
        if isinstance(data, list):
            for i, item in enumerate(data):
                if i >= 3: break
                print(f"  - {item}")
            if len(data) > 3:
                print(f"  ... (Total {len(data)} items)")
        elif isinstance(data, dict):
            for i, (key, value) in enumerate(data.items()):
                if i >= 3: break
                if isinstance(value, list):
                    print(f"  - {key}: [")
                    for j, sub_item in enumerate(value):
                        if j >= 3: break
                        sub_item_str = str(sub_item)
                        if len(sub_item_str) > 100: # Truncate long sub_items
                            print(f"    - {sub_item_str[:100]}...")
                        else:
                            print(f"    - {sub_item_str}")
                    if len(value) > 3:
                        print(f"    ... (Total {len(value)} items for key '{key}')")
                    print("  ]")
                elif isinstance(value, dict):
                    print(f"  - {key}: {{...}} (nested dictionary)") # Indicate nested dict without printing all
                else:
                    value_str = str(value)
                    if len(value_str) > 100: # Truncate long string values
                        print(f"  - {key}: {value_str[:100]}...")
                    else:
                        print(f"  - {key}: {value_str}")
            if len(data) > 3:
                print(f"  ... (Total {len(data)} keys)")
        else:
            print(f"  - {data}")
    except requests.exceptions.RequestException as e:
        print(f"Error connecting to API: {e}")
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON response: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

FEMA_api_url = "https://www.fema.gov/api/open/v2/DisasterDeclarationsSummaries"
get_json_from_api(FEMA_api_url)
```

```python
import requests
from bs4 import BeautifulSoup

# Extract paragraph text from a webpage (HTML)
# Pros: Useful for retrieving publicly available information that is human readable (websites, articles, other text content)
# Cons: Highly depends on the website's structure, relies heavily on well-structured html, otherwise
#       output will not be well-structured. some websites may have too much information resulting in unhelpful outputs
def scrape_html_text(url: str):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors

        # Read HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        print(f"Successfully scraped HTML from {url}.")
        print("First two paragraphs of text found on the page:")

        # Find and print the text of the first 5 paragraph tags
        paragraphs = soup.find_all('p')
        if paragraphs:
            for i, p in enumerate(paragraphs):
                if i >= 5: break
                print(f"  - {p.get_text(strip=True)[:300]}...") # Print first 300 chars
        else:
            print("  - No paragraph text found.")

    except requests.exceptions.RequestException as e:
        print(f"Error accessing webpage: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

html_url = "https://en.wikipedia.org/wiki/Apple"
scrape_html_text(html_url)
```
